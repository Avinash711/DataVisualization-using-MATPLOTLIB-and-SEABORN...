# MATPLOTLIB Visualization

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,10,11)
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.title('title')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.set_xlabel('x-label')
axes.set_ylabel('y-Label')
axes.set_title('Title')
axes.plot(x,y)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 =fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_xlabel('1-xlabel')
axes1.set_ylabel('1-ylabel')
axes1.set_title('Larger plot')

axes2.plot(y,x)
axes2.set_xlabel('2-xlabel')
axes2.set_ylabel('2-ylabel')
axes2.set_title('Smaller plot')


```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
plt.tight_layout()
for val in axes:
    val.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
plt.tight_layout()
axes[0].plot(x,y)
axes[1].plot(y,x)
axes[0].set_xlabel('0-xlabel')
axes[0].set_ylabel('0-ylabel')
axes[0].set_title('0-title')

axes[1].set_xlabel('1-xlabel')
axes[1].set_ylabel('1-ylabel')
axes[1].set_title('1-title')
```

```python
fig =plt.figure(figsize=(0.8,1.2),dpi=200)
ax=  fig.add_axes([0,0,1,1])
ax.plot(x,y)
#also in subplots, figsize and dpi can be set
```

```python
 fig.savefig('mydemopic.png',dpi=100)
```

```python
fig =plt.figure(figsize=(3,3))
ax=  fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x-Squared')
ax.plot(x,x**3,label='y-squared')
ax.legend(loc=0 ) #0 is for best position
#or ax.legend(loc=(0,0))
```

```python
fig =plt.figure(figsize=(8,4))
ax=  fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red',linewidth=2,linestyle='steps',marker='o',markersize=20,markerfacecolor='green',
markeredgewidth=2,markeredgecolor='yellow')
#we can use lw and ls foer line width and linestyle
#to set limit of x range and y range,lim is used
#ax.set_xlim([0,1])
#ax.set_ylim([0,100])
```

```python
fig =plt.figure(figsize=(5,3))
ax=  fig.add_axes([0,0,1,1])
ax.scatter(x,y)
```

```python
fig =plt.figure(figsize=(5,3))
ax=  fig.add_axes([0,0,1,1])
ax.hist(x,y)
```

```python
fig =plt.figure(figsize=(5,3))
ax=  fig.add_axes([0,0,1,1])
ax.boxplot([x,y])
```
